```python
print("BEFORE CLEANING:")
print(f"Total rows: {len(df)}")
print(f"Missing values:\n{df.isnull().sum()}\n")

# Make a copy to preserve original
df_clean = df.copy()

# Remove rows where critical info is missing
# A job MUST have title and company
df_clean = df_clean.dropna(subset=['job_title', 'company'])
print(f"After removing critical missing data: {len(df_clean)} rows")

# Fill salary missing with the average salary
if 'salary' in df_clean.columns:
    median_salary = df_clean['salary'].median()
    df_clean['salary'].fillna(median_salary, inplace=True)
    print("✓ Filled salary missing values with median")

# Fill location missing with 'Remote'
if 'location' in df_clean.columns:
    df_clean['location'].fillna('Remote', inplace=True)
    print("✓ Filled location missing values with 'Remote'")

# Fill job description missing
if 'job_description' in df_clean.columns:
    df_clean['job_description'].fillna('No description provided', inplace=True)
    print("✓ Filled job description missing values")

print(f"\nAFTER FILLING MISSING VALUES:")
print(f"Missing values left:\n{df_clean.isnull().sum()}")
```

```python
import pandas as pd

# Load the raw data
df = pd.read_csv('data/raw/jobs.csv')

print("="*60)
print("BEFORE CLEANING")
print("="*60)
print(f"Total rows: {len(df)}")
print(f"\nMissing values:\n{df.isnull().sum()}\n")

# Make a copy
df_clean = df.copy()

# Remove rows where critical info is missing
# A job MUST have title and location
df_clean = df_clean.dropna(subset=['title', 'location'])
print(f"After removing critical missing data: {len(df_clean)} rows")

# Fill salary missing with median
if 'med_salary' in df_clean.columns:
    df_clean['med_salary'].fillna(df_clean['med_salary'].median(), inplace=True)
    print("✓ Filled salary missing values")

# Fill description missing
if 'description' in df_clean.columns:
    df_clean['description'].fillna('No description provided', inplace=True)
    print("✓ Filled description missing values")

# Fill experience level
if 'formatted_experience_level' in df_clean.columns:
    df_clean['formatted_experience_level'].fillna('Not specified', inplace=True)
    print("✓ Filled experience level")

# Remove exact duplicates
before_dupes = len(df_clean)
df_clean = df_clean.drop_duplicates()
after_dupes = len(df_clean)
print(f"Removed {before_dupes - after_dupes} duplicate rows")

# Clean text data
df_clean['title'] = df_clean['title'].str.strip().str.title()
df_clean['location'] = df_clean['location'].str.strip().str.title()
print("✓ Cleaned text data")

print(f"\n" + "="*60)
print("AFTER CLEANING")
print("="*60)
print(f"Total rows: {len(df_clean)}")
print(f"Missing values remaining:\n{df_clean.isnull().sum()}")

# Save cleaned data
df_clean.to_csv('data/processed/jobs_cleaned.csv', index=False)
print("\n✓ Saved cleaned data to: data/processed/jobs_cleaned.csv")
```

```python
import pandas as pd

# Load cleaned data
df = pd.read_csv('data/processed/jobs_cleaned.csv')

print("="*60)
print("JOB MARKET ANALYSIS")
print("="*60)

# Top jobs
print(f"\nTOP 10 JOB TITLES")
print(df['title'].value_counts().head(10))

# Top locations
print(f"\n\nTOP 15 LOCATIONS")
print(df['location'].value_counts().head(15))

# Salary analysis
print(f"\n\nSALARY ANALYSIS")
salary_data = df['med_salary'].dropna()
print(f"Records with salary: {len(salary_data):,}")
print(f"Average: ${salary_data.mean():,.0f}")
print(f"Median: ${salary_data.median():,.0f}")
print(f"Min: ${salary_data.min():,.0f}")
print(f"Max: ${salary_data.max():,.0f}")

# Experience levels
print(f"\n\nEXPERIENCE LEVELS")
print(df['formatted_experience_level'].value_counts())

# Work types
print(f"\n\nWORK TYPES")
print(df['formatted_work_type'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv('data/processed/jobs_cleaned.csv')

# Setup styling
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Creating visualizations...\n")

# CHART 1: Top Jobs
print("1. Creating Top Jobs chart...")
top_jobs = df['title'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_jobs.plot(kind='barh', color='steelblue')
plt.xlabel('Number of Postings', fontsize=12, fontweight='bold')
plt.ylabel('Job Title', fontsize=12, fontweight='bold')
plt.title('Top 10 Most Demanded Jobs', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('visualizations/01_top_jobs.png', dpi=300, bbox_inches='tight')
plt.close()
print("   ✓ Saved: visualizations/01_top_jobs.png\n")

# CHART 2: Top Locations
print("2. Creating Top Locations chart...")
top_locs = df['location'].value_counts().head(15)
plt.figure(figsize=(12, 6))
top_locs.plot(kind='barh', color='coral')
plt.xlabel('Number of Jobs', fontsize=12, fontweight='bold')
plt.ylabel('Location', fontsize=12, fontweight='bold')
plt.title('Top 15 Job Markets by Location', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('visualizations/02_top_locations.png', dpi=300, bbox_inches='tight')
plt.close()
print("   ✓ Saved: visualizations/02_top_locations.png\n")

# CHART 3: Work Type Distribution
print("3. Creating Work Type chart...")
work_types = df['formatted_work_type'].value_counts()
plt.figure(figsize=(10, 6))
work_types.plot(kind='bar', color='mediumseagreen')
plt.xlabel('Work Type', fontsize=12, fontweight='bold')
plt.ylabel('Number of Jobs', fontsize=12, fontweight='bold')
plt.title('Jobs by Work Type', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('visualizations/03_work_type.png', dpi=300, bbox_inches='tight')
plt.close()
print("   ✓ Saved: visualizations/03_work_type.png\n")

# CHART 4: Experience Level Distribution
print("4. Creating Experience Level chart...")
experience = df['formatted_experience_level'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(experience, labels=experience.index, autopct='%1.1f%%', startangle=90)
plt.title('Job Distribution by Experience Level', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('visualizations/04_experience_level.png', dpi=300, bbox_inches='tight')
plt.close()
print("   ✓ Saved: visualizations/04_experience_level.png\n")

print("="*60)
print("✓ ALL CHARTS CREATED!")
print("="*60)
print("\nYour charts are saved in: visualizations/")
print("  • 01_top_jobs.png")
print("  • 02_top_locations.png")
print("  • 03_work_type.png")
print("  • 04_experience_level.png")
```
